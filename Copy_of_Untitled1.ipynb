```python
import pandas as pd


# Define template-based SEO content generator
def generate_seo_fields(row):
    title = row['product_title']
    desc = row['product_description']
    features = row['product_features']
    category = row['product_category']
    keywords = row.get('keywords', '')
    tags = row.get('tags', '')

    # Extract 1st two features for brevity
    feat_list = [f.strip() for f in features.split(",")[:2]]
    feat_1 = feat_list[0] if len(feat_list) > 0 else ''
    feat_2 = feat_list[1] if len(feat_list) > 1 else ''

    # SEO Title
    seo_title = f"{title} - {feat_1} & {feat_2} | {category}"

    # SEO Meta Description
    seo_meta = f"Explore {title} with {feat_1}, {feat_2}, and more. Ideal for {category} lovers. {desc[:100]}..."

    # SEO Keywords
    seo_keywords = f"{keywords}, {tags}, {feat_1.lower()}, {feat_2.lower()}, {category.lower()}"

    return pd.Series([seo_title, seo_meta, seo_keywords])

# Apply the function to generate SEO fields
df[['seo_title', 'seo_meta', 'seo_keywords']] = df.apply(generate_seo_fields, axis=1)

```

```python
def polish_meta(desc, title):
    desc = desc.strip().rstrip('.')
    return f"Explore the {title} – your go-to solution for {desc.lower()}. Ideal for modern buyers who value quality and performance."

# Apply polishing to the meta description
df["seo_meta_polished"] = df.apply(lambda row: polish_meta(row["product_description"], row["product_title"]), axis=1)

# Preview first 5 polished meta descriptions
print(df[["product_title", "seo_meta", "seo_meta_polished"]].head())
```

```python
from unsloth import FastLanguageModel
from transformers import AutoTokenizer

# Load model & tokenizer
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Phi-3-mini-4k-instruct-bnb-4bit",
    max_seq_length = 2048,
    dtype = None,  # auto-detects float16/bfloat16
    load_in_4bit = True,
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=64,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_alpha=128,
    lora_dropout=0.05,
    bias="none",
    use_gradient_checkpointing=True,
    random_state=42,
)
```

```python
import json

path = "/content/seo_dataset.jsonl"

data = []
with open(path, "r", encoding="utf-8") as f:
    for line in f:
        data.append(json.loads(line))

print(data[0])  # Preview the first example

```

```python
from datasets import Dataset

dataset = Dataset.from_list(data)
dataset
def format_row(example):
    return {
        "text": f"{example['prompt']}\n{example['completion']}"
    }

formatted_dataset = dataset.map(format_row)
formatted_dataset
```

```python
from transformers import TrainingArguments
from trl import SFTTrainer

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=formatted_dataset,
    dataset_text_field="text",
    max_seq_length=1024,
    args=TrainingArguments(
        output_dir="seo-llm-model",
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        num_train_epochs=3,
        learning_rate=2e-4,
        logging_steps=10,
        save_strategy="epoch",
        save_total_limit=2,
        report_to="none",  # no wandb
        fp16=True,
        optim="adamw_8bit"
    )
)
```

```python
trainer_stats = trainer.train()
```

```python
# Save model and tokenizer locally
trainer.save_model("seo-finetuned-checkpoint")
tokenizer.save_pretrained("seo-finetuned-checkpoint")
```
